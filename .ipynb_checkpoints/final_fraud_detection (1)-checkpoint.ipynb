1. Problem Statement

With the rapid rise in digital transactions, detecting fraudulent activities has become critical for financial institutions. Fraudulent transactions can result in huge monetary losses and damage to trust.
This project aims to develop a machine learningâ€“based web application capable of identifying potential fraudulent transactions in real time using transaction details such as type, amount, and account balances.
The goal is to create an accurate, fast, and user-friendly system that can predict fraud probability and help analysts take preventive action.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('fraud_detection.csv')
data.head()
```

```python
obj = (data.dtypes == 'object')
object_cols = list(obj[obj].index)
print("Categorical variables:", len(object_cols))
```

```python
fl = (data.dtypes == 'float')
fl_cols = list(fl[fl].index)
print("Float variables:", len(fl_cols))
```

```python
data.isnull().sum()
```

```python
data[['oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest']] = data[['oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest']].fillna(0)
data['nameDest'] = data['nameDest'].fillna('Unknown')
data['isFraud'] = data['isFraud'].fillna(data['isFraud'].mode()[0])
data['isFlaggedFraud'] = data['isFlaggedFraud'].fillna(data['isFlaggedFraud'].mode()[0])
```

```python
data.duplicated().sum()
```

```python
plt.figure(figsize=(15, 8))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=data[col])
    plt.title(f'Boxplot of {col}')
    plt.xlabel('')
plt.tight_layout()
plt.show()
```

```python
for col in numeric_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data[col] = data[col].apply(lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x)

for col in numeric_cols:
    plt.figure(figsize=(5, 5))
    sns.boxplot(x=data[col])
    plt.title(f'Capped Boxplot of {col}')
    plt.show()
```

```python
# Data Preprocessing
data['error_org'] = data['oldbalanceOrg'] - data['newbalanceOrig'] - data['amount']
data['error_dest'] = data['newbalanceDest'] - data['oldbalanceDest'] - data['amount']
data['balance_diff_org'] = data['newbalanceOrig'] - data['oldbalanceOrg']
data['balance_diff_dest'] = data['newbalanceDest'] - data['oldbalanceDest']
data['ratio_org'] = data['newbalanceOrig'] / (data['oldbalanceOrg'] + 1)
data['ratio_dest'] = data['newbalanceDest'] / (data['oldbalanceDest'] + 1)
```

```python
type_new = pd.get_dummies(data['type'], drop_first=True)
data_new = pd.concat([data, type_new], axis=1)
```

```python
# ---- Import ML Libraries ----
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score as ras
```

```python
# ---- Step 1: Additional Feature ----
data_new['log_amount'] = np.log1p(data_new['amount'])
```

```python
# ---- Step 2: Select Features ----
X = data_new.drop(['isFraud', 'type', 'nameOrig', 'nameDest'], axis=1)
y = data_new['isFraud']

print("Feature shape:", X.shape)
print("Target shape:", y.shape)
```

```python
# ---- Step 4: Train-Test Split (No SMOTE) ----
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
# ---- Step 3: Scale Selected Columns ----
scale_cols = ['amount', 'oldbalanceOrg', 'newbalanceOrig',
              'oldbalanceDest', 'newbalanceDest', 'log_amount']
scaler = StandardScaler()
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

X_train_scaled[scale_cols] = scaler.fit_transform(X_train[scale_cols])
X_test_scaled[scale_cols] = scaler.transform(X_test[scale_cols])
```

```python
# ---- Step 5: Train Logistic Regression ----
log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train_scaled, y_train)

# ---- Step 6: Evaluate ----
train_preds = log_model.predict_proba(X_train_scaled)[:, 1]
test_preds = log_model.predict_proba(X_test_scaled)[:, 1]

print("Logistic Regression")
print(f"Training AUC: {ras(y_train, train_preds):.4f}")
print(f"Validation AUC: {ras(y_test, test_preds):.4f}")
print(f"Generalization Gap: {ras(y_train, train_preds) - ras(y_test, test_preds):.4f}")
```

```python
# ---- Step 6: Define Models ----
rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=4,
    min_samples_leaf=20,
    max_features='sqrt',
    max_samples=0.8,
    criterion='entropy',
    random_state=7
)
```

```python
rf_model.fit(X_train, y_train)

# ---- Step 5: Evaluate ----
train_preds = rf_model.predict_proba(X_train)[:, 1]
test_preds = rf_model.predict_proba(X_test)[:, 1]

print("Random Forest")
print(f"Training AUC: {ras(y_train, train_preds):.4f}")
print(f"Validation AUC: {ras(y_test, test_preds):.4f}")
print(f"Generalization Gap: {ras(y_train, train_preds) - ras(y_test, test_preds):.4f}")
```

```python
import joblib

# Save the best model (Random Forest)
joblib.dump(rf_model, "rf_fraud_model.pkl")

print("\n Random Forest model saved successfully")
```
